<h1>---Programing---</h1>

```python
def Run_Program():
    while True:
        img_path_2_save = f"{root_project_path}/outputs/demonstrate_output_{random.randint(1, 100000)}.png"
        typeProblem = int(input("\n0.If you want to Exit \n 1. Using Faster RCNN to predict\n2. Using YoLoV2 to predict\n4. Using convolutuion filters"))
        if typeProblem == 0:
            print("Exiting the program.")
            break
        if typeProblem == 1:
            image_path = input("Input Image Path: ")
            img_path_2_save = predict_one_by_FasterRCNN(image_path = image_path)

            print("\033[1;31m----------Faster RCNN for your image!----------\033[0m")
            show_results(image_path, img_path_2_save)
            print(f"Predicted image saved to {img_path_2_save}")
            
        elif typeProblem == 2:
            image_path = input("Input Image Path: ")
            img_path_2_save = predict_one_by_YoLoV12(image_path = image_path)
   
            print("\033[1;31m----------YoLoV2 for your image!----------\033[0m")
            show_results(image_path, img_path_2_save)
            print(f"Predicted image saved to {img_path_2_save}")
            
        elif typeProblem == 4:
            image_or_tensor = int(input("1. if Input Image Path\n2. input tensor manually?: "))
            if image_or_tensor == 1:
                image_path = input("Input Image Path(ex: garyTest.jpg): ")
                inputTensor = mpimg.imread(image_path)
                inputTensor = transform(inputTensor)
            elif image_or_tensor == 2:
                print(f"\nPlease give me the Tensor(w*h):")
                inputTensor = supportInputTensors().unsqueeze(0)
                output_image_path = None
            else:
                print("\033[1;31mInvalid choice. Please try again.\033[0m")
                continue
                
            filter_id = int(input(f"0 if sobel_horizontal || 1 if sobel_vertical || 2 if gaussian_blur || 3 if box_blur || 4 if horizontal || 5 if vertical || 6 if diagonal_ul_lr || 7 if diagonal_ur_ll):"))
            if filter_id < 0 or filter_id >= len(id_filters):
                print("Invalid filter ID. Please try again.")
                continue
            chose_filter = id_filters[filter_id]
            
            if image_or_tensor == 1:
                output_image_path = f"{root_project_path}/outputs/convolution_{chose_filter}_{random.randint(1, 100000)}.png"
            output = convolve3cTensor(inputTensor, all_filters[chose_filter], output_image_path)
            
            print(f"\033[1;31m----------Convolution with {chose_filter} for your Input(image/tensor)!----------\033[0m")            
            if image_or_tensor == 1:
                show_results(image_path, output_image_path, is_resize=True)
                print(f"Filtered image saved to {output_image_path}")
            else:
                print(f"your input tensor({inputTensor.shape}) is:\n{inputTensor}")
                print(f"Your input tensor after convolution({inputTensor.shape}) with {chose_filter} is:\n{output}")
```

```python
Run_Program()
```
