<h1>Import Libraries</h1>

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from problem1_FasterRCNN import predict_one_by_FasterRCNN
# import subprocess
import os
# from IPython.display import clear_output
# env = os.environ.copy()
# env["MKL_THREADING_LAYER"] = "GNU"
root_project_path = os.getcwd()
print(root_project_path)
```

<h1>1. FasterRCNN</h1>

```python
def predict_one_by_FasterRCNN(image_path):
    cmd = [
        "/data2/atran16/Anaconda_ForTrain/bin/python",
        "-m", "pytorch.FasterRCNN",
        "--backbone=resnet50",
        "--load-from=fasterrcnn_pytorch_resnet50.pth",
        f"--predict={image_path}",
        f"--save-output-path={root_project_path}/outputs/demonstrate_output.png"
    ]

    subprocess.run(cmd, env=env, cwd=f"{root_project_path}/FasterRCNN")
    print("Completed!")
```

<h1>---Programing---</h1>

```python
def show_results(original_path, predicted_path):
    original_img = mpimg.imread(original_path)
    predicted_img = mpimg.imread(predicted_path)  # predicted output

    # Create subplot (1 row, 2 columns)
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # Show first image
    axes[0].imshow(original_img)
    axes[0].set_title("Original")
    axes[0].axis('off')

    # Show second image
    axes[1].imshow(predicted_img)
    axes[1].set_title("Prediction")
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()
    
def Run_Program():
    while True:
        typeProblem = int(input("\n0.If you want to Exit\n1. Using Faster RCNN to predict"))
        if typeProblem == 0:
            print("Exiting the program.")
            break
        if typeProblem == 1:
            image_path = input("Input Image Path: ")
            predict_one_by_FasterRCNN(image_path = image_path)
             # Call the function to run the Faster R-CNN predicti
            show_results(image_path, f"{root_project_path}/outputs/demonstrate_output.png")
            print("----------Faster RCNN for your image Done!----------\n")
```

```python
Run_Program()
```
